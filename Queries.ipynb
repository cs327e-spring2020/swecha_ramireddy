### February Cases from the first day of the month (2/1) to the last day of the month (2/29)

```python
%%bigquery
SELECT a.region, a.transportation_type, a._2020_02_01 as mobility_2020_02_01, b._2020_02_01 as cases_2020_02_01, a._2020_02_29 as mobility_2020_02_29, b._2020_02_29 as cases_2020_02_29 FROM `aqueous-freedom-267023.COVID_19_Cases.mobility` a
inner join `aqueous-freedom-267023.COVID_19_Cases.cases` b
on a.region = b.Country_Region
```
